```python
#Exporing the data
data = df.select_dtypes(include=['number'])

def explore_data(data):
    """
    Explores the dataset by displaying major moments, and basic statistics.

    Parameters:
        data (DataFrame): Input dataset.

    Returns:
        None
    """
    # Display major moments
    moments = pd.DataFrame({
        'Mean': data.mean(),
        'Median': data.median(),
        'Standard Deviation': data.std(),
        'Skewness': data.skew(),
        'Kurtosis': data.kurtosis()
    })
    print("Major Moments:")
    print(moments.head(50))
    #print("\n")

explore_data(data)
```

```python
def hist_plot(df):
    """
    Plot the distribution Climate changes in 1961 and 2022 across different countries

    Parameters:
    - df (DataFrame): Input dataframe containing features and target labels.

    Returns:
    - None: Displays the distribution plots.

    """
    
    
    fig = plt.figure(figsize=(14, 4))
    
    # Plot distribution over 1961 temperature
    ax1 = plt.subplot(121)
    sns.histplot(df["F1961"],color="pink",kde=True)  # using seaborn to plot histogram
    plt.title("Climate changes in 1961 ",fontsize=13)
    plt.xlabel("Temperature indicator in celsius",fontsize=13)
    plt.ylabel("Count of countries",fontsize=13)
    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
    

    # Plot distribution over 2022 temperature
    ax2 = plt.subplot(122)
    sns.histplot(df["F2022"],color="lightblue",kde=True)  # using seaborn to plot histogram
    plt.title("Climate changes in 2022 ",fontsize=13)
    plt.xlabel("Temperature indicator in celsius",fontsize=13)
    plt.ylabel("Count of countries",fontsize=13)
    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
    plt.savefig('hist.png')
    plt.show()

    return 0 
```

```python

# Function to plot Histogram
def plot_temperature_distribution(data):
    """
    plotting the Average Temperature Change Over the Years 
    """
    # Convert non-numeric values to NaN
    numeric_data = data.apply(pd.to_numeric, errors='coerce')
    mean_temperatures = numeric_data.mean()
    
    plt.figure(figsize=(18, 9))
    mean_temperatures.plot(kind='bar', color='skyblue')
    plt.title('Average Temperature Change Over the Years',fontsize=20)
    plt.xlabel('Year',fontsize=20)
    plt.ylabel('Average Temperature Change (Â°C)',fontsize=20)
    plt.xticks(rotation=90,fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig('bar_all.png')
    plt.show()
    
plot_temperature_distribution(cleaned_data.iloc[:, 1:])
```

```python
# Function to plot Line/Scatter Graph
def plot_country_temperature_change(data, country):
    """
    This plot shows the temperature change over peroid of time in India
    
    Parameters:
    - df (DataFrame): Input dataframe containing features and target labels.

    Returns:
    - None: Displays the distribution plots.

    """
    
    country_data = data[data['Country'] == country].squeeze()[1:]  # Remove the first element (Country name)
    years = range(1961, 2023)
    plt.figure(figsize=(8, 4))
    plt.plot(years, country_data, marker='h', color='blue', linestyle='-')
    plt.title(f'Temperature Change Over the Years in {country}',fontsize=12)
    plt.xlabel('Year',fontsize=12)
    plt.ylabel('Temperature Change (Â°C)',fontsize=12)
    plt.xticks(rotation=45, fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig('line.png')
    plt.show()
    
    return 
```

```python

# Function to plot Confusion Matrix/Heatmap
def plot_correlation_heatmap(data):
    correlation_matrix = selected_data.corr()
    """
    Plots a heatmap to visualize the correlation matrix of the input dataset.
    
    This function computes the correlation matrix of the input dataset and plots it as a heatmap.
    Each cell in the heatmap represents the correlation coefficient between two features.
    The correlation coefficient ranges from -1 to 1, where:
        - 1 indicates a perfect positive correlation,
        - -1 indicates a perfect negative correlation, and
        - 0 indicates no correlation between the features.
    
    Parameters:
        data (DataFrame): Input dataset.
            The dataset for which the correlation matrix will be computed and visualized.
    
    Returns:
        None
    """
    
    plt.figure(figsize=(4, 4))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Heatmap ')
    plt.tight_layout()
    plt.xticks(rotation=45, fontsize=10)
    #plt.yticks(fontsize=12)
    plt.savefig('heat.png')
    plt.show()
    
```

```python
df_d = df[[
    'Country', 'ISO3',
    'F1992', 'F2002', 'F2012', 'F2022'
]]
df_d.head()
```

```python
# choosing columns 
I = df_d.iloc[:, 2:]
list_inertia = []
X_range = range(1, 11)

#calculating inertia
for X in X_range:
    Xmeans = KMeans(n_clusters = X, n_init = 10, random_state = 84)
    Xmeans.fit(I)
    list_inertia.append(Xmeans.inertia_)
    
#plot
plt.figure(figsize = (4, 4), facecolor = "pink")
plt.plot(X_range, list_inertia, marker = "x", linestyle = '--', color='brown')
plt.xlabel('Number of Groupings (X)', fontsize=14)
plt.ylabel('Inertia', fontsize=14)
plt.title('Elbow Method for Optimal X', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.savefig('elbow.png')
plt.show()
```

```python
n_clusters = 3
kmeans = KMeans(n_clusters = n_clusters, n_init = 10, random_state = 84)
df_d['Cluster'] = kmeans.fit_predict(I)

#showings the clustered area
display(df_d)
```

```python
plt.figure(figsize = (4, 4), facecolor = "pink")
sns.scatterplot(
    data = df_d,
    x = 'F1992', y = 'F2022',
    hue = 'Cluster'
)
plt.axhline(
    0, 0, 1,
    color = "brown", linestyle = "--", alpha = 0.5, marker = "x"
)
plt.axvline(
    0, 0, 1,
    color = "brown", linestyle = "--", alpha = 0.5, marker = "x"
)

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.savefig('cluster.png')
plt.show()
```

```python

# For demonstration, let's assume we're interested in fitting data for the year range 1961 to 2022
years = [f'F{i}' for i in range(1961, 2023)]
data = df[years].mean()  # Taking the mean temperature change across countries for each year

# Drop missing values
data.dropna(inplace=True)

# Prepare the data for polynomial regression
X = np.array(range(len(data))).reshape(-1, 1)
y = data.values.reshape(-1, 1)

# Define the degree of the polynomial
degree = 3  # You can adjust this degree based on the complexity of the relationship

# Transform the features to polynomial features
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)

# Fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, y)

# Predict temperature change for future years
future_years = np.array(range(len(data), len(data) + 10)).reshape(-1, 1)
future_years_poly = poly_features.transform(future_years)
predicted_values = model.predict(future_years_poly)

plt.figure(figsize=(5, 4))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(np.concatenate((X, future_years)), np.concatenate((model.predict(X_poly), predicted_values)), color='red', label='Polynomial Fit')
plt.xlabel('Number of Years',fontsize=14)
plt.ylabel('Temperature Change',fontsize=14)
plt.title('Polynomial Regression for Temperature Change')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(fontsize=12)
plt.savefig('fitting.png')
plt.show()

```
